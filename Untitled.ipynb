```python
import h5py
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

def plot_animation(dataset, title):
    # Create a figure and axis
    fig, ax = plt.subplots()

    # Function to update the heatmap for each frame of the animation
    def update(frame):
        ax.clear()
        ax.set_title(f"{title} - Frame {frame + 1}")
        ax.imshow(dataset[:, :, frame], cmap='viridis')
    
    # Create the animation
    anim = FuncAnimation(fig, update, frames=dataset.shape[2], interval=200)
    anim.save(f'/home/samani/Documents/sim/{title}.mp4', writer='ffmpeg')

    # Show the animation
    plt.show()

# Path to the HDF file
sim_result = "/home/samani/Documents/sim/simulation.h5"

# Open the HDF file
with h5py.File(sim_result, 'r') as hdf_file:
    print(hdf_file.keys())
    
    # Access the datasets
    expected_iM_data = hdf_file["expected_iM"]
    expected_fM_data = hdf_file["expected_fM"]
    expected_ibM_data = hdf_file["expected_ibM"]
    expected_bM_data = hdf_file["expected_bM"]
    expected_cells_anker_all = hdf_file["expected_cells_anker_all"]
    #print(expected_iM_data[:])

    # Plot animations for different datasets
    plot_animation(expected_iM_data, "expected_iM")
    plot_animation(expected_fM_data, "expected_fM")
    plot_animation(expected_ibM_data, "expected_ibM")
    plot_animation(expected_bM_data, "expected_bM")
    #plot_animation(expected_cells_anker_all, "expected all")

```
